# F1 Data Exploration

```python
import fastf1

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
list_of_dfs = []
for year in range(2014, 2025):
    print(year)
    quali = fastf1.get_session(year, "Singapore Grand Prix", "Q")
    quali.load()

    df = quali.results
    df["year"] = year

    list_of_dfs.append(df)
```

```python
quali_df = pd.concat(list_of_dfs)
quali_df.head()
```

```python
quali_df.columns
```

```python
quali_df[["FullName", "Position", "ClassifiedPosition", "GridPosition", "Q1", "Q2", "Q3", "year"]]\
    .groupby("")
```
